# Get name-based data, each combined into (set 1, set 2, set 3)

# 
find_idx ----> return a list of T/F array 
set_loc  ----> for each test_name ('Mental','Physical' ...) use find_idx() to get their index location. Then access the the data frame of such location, creating a name-based dataframe

The name based data frame were then convert into its own set category
- mental demand (set 1, set 2, set 3)

```python
# return the indices of the desired NASA test name, as pd.Index() return an array of T/F
# we return the indice when the condition = true
num_sets = 3
# converting the T/F array to a list of index which will then be used in set_loc
def find_idx(data):
    ls =[]
    for i in range(0,len(data)):
        if data[i]:
            ls.append(i)
    return ls
        

def set_loc(test_name):
    tst_name = pd.Index(combine['Test name']).get_loc(test_name) # get the location of desired set name, e.g. 'Mental command
    name_idx = find_idx(tst_name)
    name_frame = combine.iloc[name_idx,:12]
    name_frame.reset_index(drop=True, inplace = True)

    sets = ['set 1', 'set 2', 'set 3']
    set_frame = [[] for i in range(0,num_sets)]
    for i in range(0, len(sets)):
        set_name  = pd.Index(name_frame['Sets']).get_loc(sets[i])
        set_idx = find_idx(set_name)
        set_frame[i] = name_frame.iloc[set_idx,:12]
        set_frame[i].reset_index(drop = True, inplace = True)

    return set_frame
    
set_loc('Physical demand ')
# set_loc('Mental demand')
```

```python
# mental_d = set_loc('Mental demand')

# This function provide the according name as well as set separation of the data
# data[0] is mental demand with all 3 sets of data, 
# data[0][0] = set 1
# data[0][1] = set 2
def nasa_name_set():
    names = pd.Index(tlx_1['Test'])
    data = [[] for i in range(0, len(names))]
    for i in range(0,len(names)):
        data[i] = set_loc(names[i])
    return data


def plt_com(data_name= nasa_name_set()):
    
    data = [[] for i in range(0, len(data_name))]
    
    for i in range(0, len(data_name)):
        s1_data = np.asarray(data_name[i][0].iloc[:9,2:]).flatten()
        s2_data = np.asarray(data_name[i][1].iloc[:9,2:]).flatten()
        s3_data = np.asarray(data_name[i][2].iloc[:9,2:]).flatten()
        data[i] = [s1_data,s2_data,s3_data]

    return data
plt_com()
```

```python
# initial frame will return a data frame with 3 rows, each representing each set
# ----> has 8 columns, one for each nasa qus
# all data within has been previously flattened already

# goal: reshape the data frame into
# ----> keep the 8 columns, but unpack the array within each element
# ----> so turn set 1 into a pd series then concat it with set 2 & 3

def nasa_final(dic = dic, name_lst= pd.Index(tlx_1['Test'])):
    initial_frame = pd.DataFrame(dic)
    
    s1_data = [[] for i in range(len(name_lst))]
    s2_data = [[] for i in range(len(name_lst))]
    s3_data = [[] for i in range(len(name_lst))]

    set_1_final = [[] for i in range(0,len(name_lst))]
    set_2_final = [[] for i in range(0,len(name_lst))]
    set_3_final = [[] for i in range(0,len(name_lst))]
    final_col = [[] for i in range(0,len(name_lst))]
    
    for i in range(0,len(name_lst)):
        s1_data[i] = pd.Series(initial_frame.loc[0,name_lst[i]])
        s2_data[i] = pd.Series(initial_frame.loc[1,name_lst[i]])
        s3_data[i] = pd.Series(initial_frame.loc[2,name_lst[i]])

        # provide a column with set information as index
        idx_s1 = pd.Series(['set 1' for j in range(0,len(s1_data[0]))])
        idx_s2 = pd.Series(['set 2' for j in range(0,len(s2_data[0]))])
        idx_s3 = pd.Series(['set 3' for j in range(0,len(s3_data[0]))])

        print(s1_data[i])
        # print(idx_s1)
        s1_data[i].index = idx_s1
        s2_data[i].index = idx_s2
        s3_data[i].index = idx_s3

        # print(s1_data[i])

        set_1_final[i] = s1_data[i].reset_index(drop = False)
        set_2_final[i] = s2_data[i].reset_index(drop = False)
        set_3_final[i] = s3_data[i].reset_index(drop = False)

        set_1_final[i].columns =['Sets',name_lst[i]]
        set_2_final[i].columns =['Sets',name_lst[i]]
        set_3_final[i].columns =['Sets',name_lst[i]]

        #
        final_col[i] = pd.concat([set_1_final[i], set_2_final[i], set_3_final[i]], axis = 0)
        
        # final_col[i] = pd.concat([s1_data[i], s2_data[i], s3_data[i]], axis = 0)

    return final_col
```

Concating the 8 questions together into one big data frame

```python
# a = final column
path = 'C:/Users/evach/Documents/MSc/Diss_data/3_set_baseline/All/Human_NASA/'

def nasa_combine (a = nasa_final()):
    for i in range(1,len(pd.Index(tlx_1['Test']))):
        a[i].drop('Sets',axis=1,inplace = True)
    end_nasa = pd.concat([a[0],a[1],a[2],a[3], a[4], a[5], a[6],a[7]], axis = 1)

    end_nasa.to_csv(path+'manova friendly'+'.csv')

nasa_combine()
```
